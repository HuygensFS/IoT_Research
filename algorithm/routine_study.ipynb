```python
##Part 1
import matplotlib.pyplot as plt
import numpy as np
#%matplotlib inline
import pandas as pd

new_day_marker = 18000 # Number of seconds until next sequence of events
n_limit = 20 # Number of events to analyse per day start
dif_sec = 60 #define the time between events
file_name = 'event_data.json'

df = pd.read_json(file_name,convert_dates=False,convert_axes=False)#,typ='series')
time = df['timestamp']
dp = df['device_position']

list_num = []
count = 0
for i in range(len(time)):
    if(i != 0):
        if((time[i] - time[i-1])> new_day_marker):
            list_num.append([])
            for t in range(0, n_limit):
                list_num[count].append(i+t)
            count = count + 1
print(count)
##Part 2
sequences = []
codes = []
count = 0

for i in range(len(list_num)):
    checker = []
    for j in range(len(list_num[i])):
        #Compare times and vote
        if(j == 0):
            checker = []
            sequences.append([])
            sequences[count].append(list_num[i][j])
            codes.append([])
            codes[count].append(dp[list_num[i][j]])
            checker.append(dp[list_num[i][j]])
        else:
            if(time[list_num[i][j]] - time[list_num[i][j-1]] < dif_sec and dp[list_num[i][j]] not in checker):
                sequences[count].append(list_num[i][j])
                codes[count].append(dp[list_num[i][j]])
                checker.append(dp[list_num[i][j]])
            else:
                checker = []
                sequences.append([])
                codes.append([])
                count = count + 1
                sequences[count].append(list_num[i][j])
                codes[count].append(dp[list_num[i][j]])
                checker.append(dp[list_num[i][j]])
    count = count + 1  

##Part 3
codes_list = []
codes_vote = []
count = 0
for i in range(len(codes)):
    a = 0
    for j in range(len(codes_list)):
        if np.array_equal(codes[i],codes_list[j]):
            codes_vote[j] = codes_vote[j] + 1
            a = a+1
    if(a == 0):
        if(i > 0):
            count = count + 1
        codes_list.append([])
        #codes_vote.append([])
        for t in range(len(codes[i])):
            codes_list[count].append(codes[i][t])
        codes_vote.append(1)     
total_list = zip(np.transpose(codes_list), np.transpose(codes_vote))
total_list = list(total_list)
#print(total_list)


##Only post data processing
final_list = []
vote_count = []
count = 0
for i in range(len(total_list)):
    if(len(total_list[i][0]) > 2):
        for j in range(len(total_list[i][0])):
            aux = total_list[i][0]
            if(j > 0):
                final_list.append([])
                if(aux[j-1] < aux[j]):
                    final_list[count].append([aux[j-1],aux[j]])
                else:
                    final_list[count].append([aux[j],aux[j-1]])
                final_list[count].append(total_list[i][1])
                count = count + 1
    elif(len(total_list[i][0]) == 2):
        final_list.append([])
        if(total_list[i][0][0] < total_list[i][0][1]):
            final_list[count].append(total_list[i][0])
        else:
            final_list[count].append([total_list[i][0][1],total_list[i][0][0]])
        final_list[count].append(total_list[i][1])
        count = count + 1
    else:
        final_list.append([])
        final_list[count].append(total_list[i][0])
        final_list[count].append(total_list[i][1])
        count = count + 1
#print(final_list)

new_list = []
for i in range(len(final_list)):
    ver = 0
    for j in range(len(new_list)):
        if(final_list[i][0] == new_list[j][0]):
            new_list[j][1] = new_list[j][1] + final_list[i][1]
            ver = 1
    if(ver == 0):
        new_list.append(final_list[i])
    
print(new_list)    
with open('out.txt', 'w') as f:
    print(new_list, file=f)
    
calib_list = []
for i in range(len(new_list)):
    if(new_list[i][1] > 1):
        if(len(new_list[i][0]) < 2):
            calib_list.append([new_list[i][0][0],new_list[i][0][0]])
        else:
            calib_list.append(new_list[i][0])
print(calib_list)
```
